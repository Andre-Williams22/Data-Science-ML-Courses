# Matplotlib Example Workflow 

```python
# 0. import matplotlib 
import matplotlib.pyplot as plt 

# 1. prepare data
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40 ,50]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))
#3. plot data
ax.plot(x,y)

#4. customize plot 
ax.set(title="Simple Plot", xlabel='x-axis', ylabel='y-axis')


#5. save & show 
fig.savefig("../images/sample-plot.png")
```

## Two options for subplots

```python
# subplot option 1 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.rand(100));
```

```python
# Subplots option 2 
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(100))
```

# Pyplot vs Matplotlib OO method
* When plotting something quickly, use pyplot method 
* when plotting something advance use the 00 method

```python
over_50 = hd[hd["age"] > 50]
over_50.head()
```

```python
# pyplot method 
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=[10,6]);
```

```python
over_50.target.values
```

```python
# OO Method mixed with pyplot method
fix, ax = plt.subplots(figsize=[10, 6])
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
#ax.set_xlim([45, 100])
```

```python
# object oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50['chol'],
                     c=over_50['target']);

# customize 
ax.set(title='heart disease and cholestoral levels',
      xlabel='age',
      ylabel='cholestorol');

# add a legend 
ax.legend(title='Target');

# add a horizontal line 
ax.axhline(over_50["chol"].mean(), linestyle='--')
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach 
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50['chol'],
                     c=over_50['target'])
# customize 
ax0.set(title="Heart disease and cholesterol levels",
       xlabel='Age',
       ylabel='cholesterol')
# add a legend to ax0
ax0.legend(title="Target")
# Add a meanline
ax0.axhline(y=over_50['chol'].mean(),
           linestyle="--");

scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
ax1.set(title="Heart disease and Max Heart Rate", 
       xlabel='age',
       ylabel='Max Heart Rate')

ax1.legend(title='Target')
ax1.axhline(y=over_50['thalach'].mean(), linestyle="--");

# Add title to figure 
fig.suptitle("Heart Disease Analysis", fontsize=20,fontweight='bold')
```
