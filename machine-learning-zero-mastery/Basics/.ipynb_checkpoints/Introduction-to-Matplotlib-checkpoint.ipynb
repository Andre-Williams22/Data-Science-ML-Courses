# Matplotlib 
Built on numpy, integrates pandas, visualizes data 

1. Create Data
2. Create plot 
3. plot data(axes on figure)
4. Customize plot
5. Save/share plot

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np 
```

```python
x = np.array([1, 2, 4, 6, 8, 10])
y = np.array([10, 20, 30, 40, 50, 70])
```

```python
plt.plot(x,y)
```

```python
# recommended method 
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 250, 300, 400, 500]) # add some data
```

# Matplotlib Example Workflow 

```python
# 0. import matplotlib 
import matplotlib.pyplot as plt 

# 1. prepare data
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40 ,50]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))
#3. plot data
ax.plot(x,y)

#4. customize plot 
ax.set(title="Simple Plot", xlabel='x-axis', ylabel='y-axis')


#5. save & show 
fig.savefig("../images/sample-plot.png")
```
