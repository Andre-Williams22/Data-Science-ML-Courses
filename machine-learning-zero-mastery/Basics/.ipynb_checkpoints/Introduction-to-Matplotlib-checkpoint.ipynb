# Matplotlib 
Built on numpy, integrates pandas, visualizes data 

1. Create Data
2. Create plot 
3. plot data(axes on figure)
4. Customize plot
5. Save/share plot

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np 
```

```python
x = np.array([1, 2, 4, 6, 8, 10])
y = np.array([10, 20, 30, 40, 50, 70])
```

```python
plt.plot(x,y)
```

```python
# recommended method 
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 250, 300, 400, 500]) # add some data
```

# Matplotlib Example Workflow 

```python
# 0. import matplotlib 
import matplotlib.pyplot as plt 

# 1. prepare data
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40 ,50]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))
#3. plot data
ax.plot(x,y)

#4. customize plot 
ax.set(title="Simple Plot", xlabel='x-axis', ylabel='y-axis')


#5. save & show 
fig.savefig("../images/sample-plot.png")
```

## Making figures with Numpy Arrays

We want:
* Line plot
* scatter plot 
* bar plot 
* histogram
* subplots

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100)
#y = np.linspace(0, 20, 100)
```

```python
#plot the data
fig, ax = plt.subplots()
ax.plot(x,x**2)
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# make a plot from dictionary
nut_butter_prices = {"Almond butter": 10, 
                    "Peanut butter": 8,
                    "Cashew butter": 20}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Dre's nut butter store", ylabel='Price ($)');
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
x = np.random.randn(1000)

fig,ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# subplot option 1 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.rand(100));
```

```python
# Subplots option 2 
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(100))
```
